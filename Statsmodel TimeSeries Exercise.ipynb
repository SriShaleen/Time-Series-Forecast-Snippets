```python
df['SMA_12'] = df['Energy'].rolling(window=12).mean()
```

```python
df.plot()
```

```python
df.head(20)
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

span = 12
alpha = 2/(span+1)

df['EWMA_12'] = SimpleExpSmoothing(df['Energy']).fit(smoothing_level=alpha, optimized=False).fittedvalues.shift(-1)
```

```python
df.head(15)
```

```python
df.plot(figsize=(12,8))
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

df['TES_12'] = ExponentialSmoothing(df['Energy'], trend='mul', seasonal='mul', seasonal_periods=12).fit().fittedvalues
```
