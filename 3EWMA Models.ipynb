### EWMA : Exponential Weighted Moving Average.

```python
# Disadvantages of Simple Moving Average were:

# In SMA, entire model is constrained to same window size.
# Basic SMA(i.e Rolling mean) has some weakness like small windows lead to more noise, rather than signal.
# Moreover, with basic SMA we will always lag by the size of the window.
# Also, basic SME will never reach to full peak or valley of the data due to averaging.
# Moreover, SMA not really inform you abt possible future behaviour, all it does is describes trends in your data
# Also, extreme historical values can skew SMA results significantly.
```

```python
# So, EWMA helps fix these issues of SMA and gives recent data to be more weighted than older data.
# Thus, EWMA will allow us to reduce the lag effect from SMA, and it will put more weight on values that occured more recently.
```

### Firstly, lets see the simple moving average (or rolling mean) we already learnt:

```python
airline['6_sma_passengers'] = airline['Passengers'].rolling(window=6).mean()
airline['12_sma_passengers'] = airline['Passengers'].rolling(window=12).mean()
```

```python
airline[['Passengers','6_sma_passengers','12_sma_passengers']].plot(figsize=(12,5))
```

### Lets check the EWMA using ewm() function:

```python
airline['EWMA-12'] = airline['Passengers'].ewm(span=12).mean()
```

```python
airline[['Passengers','EWMA-12','12_sma_passengers']].plot(figsize=(10,8))
```
