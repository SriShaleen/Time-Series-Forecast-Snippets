```python
# There are 2 types of ETS models: Additive model and multiplicative model. 
# We apply additive model, when it seems the trend is more linear, and the seasonality and trend component seems to be constant over time. eg. Every year we add 10K passengers.
# We apply multiplicative model, when we observe increase/decrease of trend non-linearly. eg: Each year we double the passengers from previous year. 
```

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
airline = pd.read_csv("C:\\Users\\shaleens\\Desktop\\TimeSeriesUdacity\\TS Udemy Vids\\Time Series 2\\Airlines.csv", index_col=0, parse_dates=True)
```

```python
airline.head()
```

```python
airline.dropna(inplace=True)
```

```python
airline.plot()
```

```python
# Note: we can observe that the trend is increasing at a slightly higher rate
# than linear. So, we will be using multiplicative model.
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(airline['Passengers'], model='multiplicative')
type(result)
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 12,5

result.plot();
```
